```python
import matplotlib.pyplot as plt
import numpy as np

fig = plt.figure()
ax = plt.subplot(projection="3d")


x, y, z = np.meshgrid(
    np.arange(-0.8, 1, 0.2), np.arange(-0.8, 1, 0.2), np.arange(-0.8, 1, 0.8)
)

u = y * z
v = x * z
w = 2 * x

ax.quiver(x, y, z, u, v, w, length=0.2)

plt.show()
```
